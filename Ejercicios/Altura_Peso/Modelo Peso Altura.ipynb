```python
# Importamos librería pandas para la gestión de datos
import pandas as pd

# Leemos el archivos CSV, fuete de los datos para entrear el modelo
datos = pd.read_csv('personas_colombianas_altura_peso.csv')
```

```python
# Muestra información de los datos cargados
datos.info()
```

```python
# Se muestran los primeros datos
datos.head()
```

```python
# Se importa la librería seaborn para la visualización de datos
import seaborn as sb

# Se crea un gráfico con los datos de altura y peso
sb.scatterplot(x="altura", y="peso", data = datos, hue = "peso", palette="coolwarm")
```

```python
# Se asignan los datos: Características (X), etiquetas (y)
x = datos["altura"]
y = datos["peso"]
```

```python
# Se hace el Reshape de los datos para poder procesarlos
x_procesado = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
# Se importa la librería Sklearn para la creación del modelo
from sklearn.linear_model import LinearRegression

# Se crea el objeto del modelo de regresión lineal
modelo = LinearRegression()
```

```python
# Se realiza el Entreamiento del modelo
modelo.fit(x_procesado, y_procesada)
```

```python
# Se prueba del modelo #

mi_altura = 170 # Dato de entrada

# Se predice el peso a partir de la altura
prediccion_peso = modelo.predict([[mi_altura]])

# Se muestra el resultado de la predicción
print(f"{mi_altura} cm de altura, {prediccion_peso[0][0]} kg de peso")
```

```python
# Se revisa el score del modelo (Cercano a 1 es mejor)
modelo.score(x_procesado, y_procesada)
```
